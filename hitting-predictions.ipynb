# MLB Hall of Fame

Usng Lahman's Baseball Database to predict which players will be inducted into the MLB Hall of Fame. Predictive analytics of hall of fame members.

#### Dataset sources:   
Sean Lahman, Lahman's Baseball Database: [MLB_Kaggle](https://www.kaggle.com/seanlahman/the-history-of-baseball/data)

We limit ourselves to the following .csv files in this analysis:
- `all_star.csv`: All-Star appearances
- `batting.csv` and `pitching.csv`: Batting data and pitching data, used to evaluate player value for HOF candidacy
- `player_award_vote.csv`: MVP, Cy Young, and other votes
- `hall_of_fame.csv`: Response variable, induction into HOF
- `player.csv`: Dataset of player information to match up names, etc.

```python
# base python packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn pre-processing
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

# sklearn model selection
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.pipeline import Pipeline
from sklearn import metrics

# sklearn models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# block warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# import datasets
players = pd.read_csv('clean_data/players.csv')
hit = pd.read_csv('clean_data/batting.csv')
pitch = pd.read_csv('clean_data/pitching.csv')

# change indexes to player_id
players.set_index('player_id', inplace = True)
hit.set_index('player_id', inplace = True)
pitch.set_index('player_id', inplace = True)

# field = pd.read_csv('clean_data/pitching.csv')
# field.set_index('player_id', inplace = True)

players.head()
```

```python
# sort by VOTE% so we have the highest option for each player, then reduce indices
players.sort_values(by = ['VOTE%'], ascending = False, inplace = True)
players = players[~players.index.duplicated(keep = 'first')]
```

## Preparing data for model fitting
Separating active from inactive, separating vote and induction, separating name to leave just standard features.

```python
# player data only
player_bio = players[['name', 'active', 'seasons_played']]

# active players
players_active = players[players.active == 1].copy()
players_active.drop(labels = ['active', 'VOTE%', 'inducted'], axis = 1, inplace = True)

# hall-of-fame eligible
hof_eligible = players[(players.active == 0) & (players.seasons_played >= 10)].copy()
hof_eligible.drop(labels = ['name', 'active'], axis = 1, inplace = True)

# voting results - only highest vote % for each
hof_vote = hof_eligible[['VOTE%']].copy()
hof_induction = hof_eligible[['inducted']].copy()

# reducing to just features
hof_eligible.drop(labels = ['VOTE%', 'inducted'], axis = 1, inplace = True)
hof_eligible.head()
```

## Classifier runs to check for data quality
Used before creating final dataset.

Basic classifier (decision tree) based on these variables alone.

```python
# preparing dataset
X = hof_eligible.copy()
y = hof_induction.copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# creating model
model = DecisionTreeClassifier(max_depth = 5)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# model fit is "good," but we also have much more 0's to begin with, so this isn't a real improvement for us
print(hof_induction.groupby('inducted')['inducted'].count())
(3165-115)/3165
```

## Creation of final test/training datasets
Combining batting and pitching data to certain players for a full dataset.

```python
# joining the hall of fame inducted options first to simplify work later
hitting_full = pd.concat([hit, hof_vote, hof_induction], axis = 1, sort = True)
hitting_full = hitting_full[hitting_full['ab'] >= 0] # remove rows that do not match up with data
```

```python
# pitchers as players in both hitting and pitching datasets
pitchers = pitch.merge(hitting_full[['ab', 'VOTE%', 'inducted']], how = 'left', left_index = True, right_index = True)
pitchers.head()
```

```python
# verify that this does not remove pitcher rows
len(pitch) == len(pitchers)
```

Now, let's remove pitchers who had more at-bats than they did outs recorded; this suggests they were hitters called into a game, for example. As we can see, this reduces the number of players.

```python
# pitchers must have gotten more outs than at-bats/opportunities to make outs
pitchers = pitchers[pitchers.ab <= pitchers.ipouts]
pitchers.fillna(value = 0, axis = 1, inplace = True)
pitchers.drop(labels = 'ab', axis = 1, inplace = True)
print(len(pitchers))
pitchers.head()
```

Now, using this, we can find the hitters only, as those not with an index in pitchers.

```python
# batters as players who are not pitchers
hitters = hitting_full.merge(pitchers[['ipouts']], how = 'left', left_index = True, right_index = True)
hitters = hitters[(~(hitters.ipouts > 0)) | (hitters.ipouts < hitters.ab)] # no values of ipouts or more ab's
hitters.drop(labels = ['ipouts'], axis = 1, inplace = True)
hitters.head()
```

```python
# verify that we have the right total number of players
len(hitting_full) == len(hitters) + len(pitchers)
```

Concatenate these to their `hof_eligible` tables to only include players who are eligible (in our model later, we will on predict on eligible future players for simplicity). Moreover, separating so we only have one response variable for each. Later, we will only used the induction first because it is the simpler variable.

```python
# full dataframes
hitting_df = hof_eligible.merge(hitters, left_index = True, right_index = True)
pitching_df = hof_eligible.merge(pitchers, left_index = True, right_index = True)

# dataframes with VOTE%
hitting_vote = hitting_df.drop(labels = ['inducted'], axis = 1).copy()
pitching_vote = pitching_df.drop(labels = ['inducted'], axis = 1).copy()

# dataframes with induction
hitting_hof = hitting_df.drop(labels = ['VOTE%'], axis = 1).copy()
pitching_hof = pitching_df.drop(labels = ['VOTE%'], axis = 1).copy()
```

## Scaling and initial tests of data - hitting
Re-preparing data and retraining models to optimize for fit. For the first section, let's start with a LogisticRegression model, add in DecisionTree models, and try an SVM as well based on the categorical (binary) type of data.

We drop the `seasons_played` variable since the granular-level data is more relevant to us.

```python
# setting a random state
rs = 19
```

```python
# preparing dataset, removing the pitching and negative statistics
hitting_hof.drop(labels = ['CYA', '3000K', '300W', '300SV', 'so', 'g_idp', 'cs', 'SAC'], axis = 1, inplace = True)

# train/test
X = hitting_hof.copy()
X['All-Star'] = X['All-Star']/X['seasons_played']
X.drop(labels = ['seasons_played', 'inducted'], axis = 1, inplace = True)
y = hitting_hof[['inducted']].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = rs)

# creating model
model = LogisticRegression(random_state = rs)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# standard scaling, neglecting our cols that do not need scaling due to their binary or percentage nature
hitting_cols = list(X_train)
hitting_scalable = hitting_cols[6:]

# partially standard scaling
def standardscaling(X_train, X_test, scalable):
    scaler = StandardScaler()
    X_train[scalable] = scaler.fit_transform(X_train[scalable])
    X_test[scalable] = scaler.transform(X_test[scalable])
    return X_train, X_test, scaler

X_train, X_test, hit_scaler = standardscaling(X_train, X_test, hitting_scalable)
X_train.head()
```

```python
# rerunning logistic regression 
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# cross-validation
scores = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 10)
print('CV accuracy: %s' % scores)
print('CV accuracy range: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# plot learning curve
pipe_lr = Pipeline([('clf', model)])
train_sizes, train_scores, test_scores = \
                learning_curve(estimator = pipe_lr,
                        X = X_train, y = y_train,
                        train_sizes = np.linspace(0.2, 1.0, 8), 
                        cv = 10)

# learning curve ranges
def learningcurve_ranges(train_scores, test_scores):
    train_mean = np.mean(train_scores, axis = 1)
    train_std = np.std(train_scores, axis = 1)
    test_mean = np.mean(test_scores, axis = 1)
    test_std = np.std(test_scores, axis = 1)
    return train_mean, train_std, test_mean, test_std

# plotting learning curve
def plot_learningcurve(train_sizes, train_scores, test_scores, model_name):
    train_mean, train_std, test_mean, test_std = learningcurve_ranges(train_scores, test_scores)
    plt.plot(train_sizes, train_mean, 'b', marker='o')
    plt.plot(train_sizes, test_mean, 'r', marker='o')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color = 'red', alpha = 0.1)
    
    plt.title('Learning curve, {}'.format(model_name))
    plt.legend(('training', 'test'), fancybox = True)
    plt.show()

plot_learningcurve(train_sizes, train_scores, test_scores, 'LogisticRegression')
```

We get a slight improvement from scaling these values, especially since standard scaling is comparing to averages in the groups. Cross-validation suggests we are not overfitting too heavily with our logisitic regression, but the learning curve suggests otherwise. Let's look into dimensionality reduction.

## PCA
Dimensionality reduction of the data

```python
# let's determine optimal number of components
pca_model = PCA(n_components = len(hitting_cols)) # all components
X_PCA = pca_model.fit_transform(X_train)

variance = pca_model.explained_variance_ratio_ #variance ratios                                        

# cumulative sum of variance, for the first [n] features
cv = np.cumsum(np.round(pca_model.explained_variance_ratio_, decimals = 3)*100)
```

```python
# plot cv vs features
plt.plot(cv, c = 'black')
plt.xlabel('number of features')
plt.ylabel('cumulative sum variance')
plt.xticks(np.arange(0, len(hitting_cols), 1))
plt.title('cv of pca = {}'.format(len(hitting_cols)))
plt.axhline(y = 99, color = 'purple', alpha = .75)
plt.axhline(y = 98, color = 'green', alpha = .75)
plt.axhline(y = 95, color = 'orange', alpha = .75)
plt.axhline(y = 90, color = 'yellow', alpha = .75)
ax = plt.axes()        
ax.xaxis.grid(True, alpha = 0.5) 
plt.show()
```

Our key takeaways from PCA are that we can explain 90% of variance with 2 features, 95% with 3-4 features, 98 with 5, and 99 with 7 features. This means there is significant potential to reduce dimensionality in our model inputs.

This isn't really surprising, either - a lot of baseball statistics should be correlated by the way they are recorded (plus, people only look at the same couple stats usually...). Let's look at 95% variance with 4 features (we can see the learning curves also look best for this)

```python
# PCA models
PCA_hit = PCA(n_components = 6) # 98%

# fit PCA, 98%
X_train_PCA = PCA_hit.fit_transform(X_train)
X_test_PCA = PCA_hit.transform(X_test)

# rerunning logistic regression 
model.fit(X_train_PCA, y_train)
print('Model score: {}'.format(model.score(X_test_PCA, y_test)))

# cross-validation
scores = cross_val_score(estimator = model, X = X_train_PCA, y = y_train, cv = 10)
print('CV accuracy: %s' % scores)
print('CV accuracy range: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))

# plot learning curve
pipe_lr = Pipeline([('clf', model)])
train_sizes, train_scores, test_scores = \
                learning_curve(estimator = pipe_lr,
                        X = X_train_PCA, y = y_train,
                        train_sizes = np.linspace(0.2, 1.0, 8), 
                        cv = 10)
plot_learningcurve(train_sizes, train_scores, test_scores, 'LogisticRegression + 95% PCA')
```

## Modeling and final predictions - hitters
We will combine our standard scaling with PCA-98% as our new training set for the model itself, in our hitters. Let's use an SVM model as well, comparing to our Logistic Regression, Decision Trees, and Random Forests. However, the DecTree or Random Forest makes the most sense from our modeling perspective given the nature of the choices that are made in the human-based voting.

Schema for `classifier_results` to compare different predictive algorithms copied from HW5 in [CIS 545: Big Data Analytics](https://www.cis.upenn.edu/~cis545/)

```python
# preparing dataset
X = hitting_hof.copy()
X.drop(labels = ['seasons_played', 'inducted'], axis = 1, inplace = True)

y = hitting_hof[['inducted']].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = rs)

# classifier results dictionary
classifier_results = []
```

```python
# standard scaling and PCA
X_train, X_test, hit_scaler = standardscaling(X_train, X_test, hitting_scalable)
X_train_PCA = PCA_hit.fit_transform(X_train)
X_test_PCA = PCA_hit.transform(X_test)
```

```python
# helper function for classification
def model_score(model):
    model.fit(X_train_PCA, y_train)
    test_score = model.score(X_test_PCA, y_test)
    return test_score
```

```python
# baseline: 96.19%
print(y.groupby('inducted')['inducted'].count())
classifier_results.append({'Classifier': 'Everything = 0', 'Score': (1966 - 75)/1966})
(1966 - 75)/1966
```

```python
# decision trees for various heights - we only have 4 PCs
for i in range(1, 6, 1):
    dt_model = DecisionTreeClassifier(max_depth = i, random_state = rs)
    classifier_results.append({'Classifier': 'DecTree', 'Depth': i, 'Score': model_score(dt_model)})

for i in [10, 15, 20]:
    rf_model = RandomForestClassifier(n_estimators = i, random_state = rs)
    classifier_results.append({'Classifier': 'RandomForest', 'Score': model_score(rf_model), \
                               'Count': len(rf_model.estimators_)})
    
# model fitting as before
SVM_model = SVC(random_state = rs)
classifier_results.append({'Classifier': 'SVC', 'Score': model_score(SVM_model)})

lr1_model = LogisticRegression(penalty = 'l1', random_state = rs)
classifier_results.append({'Classifier': 'LogReg-L1', 'Score': model_score(lr1_model)})

lr2_model = LogisticRegression(penalty = 'l2', random_state = rs)
classifier_results.append({'Classifier': 'LogReg-L2', 'Score': model_score(lr2_model)})
```

Now let's try some more advanced models - boosts, baggings. We'll use a baseline of 10 estimators.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier

## Bagging classifier Decision Tree, depths varied
for i in range(3, 6, 1):
    dt_model = DecisionTreeClassifier(max_depth = i, random_state = rs) 
    
    dt_bag = BaggingClassifier(base_estimator = dt_model, n_estimators = 10, random_state = rs)
    classifier_results.append({'Classifier': 'Bag-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

    dt_bag = AdaBoostClassifier(base_estimator = dt_model, n_estimators = 10, random_state = rs)
    classifier_results.append({'Classifier': 'Boost-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

## Boost classifier SVM
svm_boost = AdaBoostClassifier(base_estimator = SVM_model, algorithm = 'SAMME', n_estimators = 10, random_state = rs)
classifier_results.append({'Classifier': 'Boost-SVM', 'Score': model_score(svm_boost), 'Count': len(svm_boost.estimators_)})
```

Let's view our classifiers and select a model.

```python
# output classifiers as a dataframe
hitting_classifiers = pd.DataFrame(classifier_results)

# baseline score
baseline = hitting_classifiers.iloc[0, 3]
hitting_improved = hitting_classifiers[hitting_classifiers['Score'] > baseline]

# models
hitting_improved.sort_values(by = 'Score', inplace = True, ascending = False)
hitting_improved[0:20]
```

A Random Forest or Boosted Decision Tree depth 5 seems like the simplest yet high-performing option. Moreover, considering that there is less of a continuum on Hall of Fame selection and rather different cutoff considerations per se that the writers may use (ex: hitting ability, getting on base), let's consider using the Tree first.

```python
tree_select = []
dt_model = DecisionTreeClassifier(max_depth = 5, random_state = rs)

for i in range(1, 10): 
    dt_bag = BaggingClassifier(base_estimator = dt_model, n_estimators = i, random_state = rs)
    dt_boost = AdaBoostClassifier(base_estimator = dt_model, n_estimators = i, random_state = rs)
    tree_select.append({'Classifier': 'Bag-DecTree', 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})
    tree_select.append({'Classifier': 'Boost-DecTree', 'Score': model_score(dt_boost), 'Count': len(dt_boost.estimators_)})

tree_df = pd.DataFrame(tree_select)
tree_df.drop_duplicates(subset = ['Score', 'Count'], inplace = True)
print(tree_df.groupby(['Classifier'])['Score'].max())

# maximum is DecisionTree Bag
tree_df
```

## Final model: hitters
For the sake of parsimony and model ability, we select the standard Decision Tree depth 3 since it does not do any worse than the bagged models. Now we can look at which are mispredicted. We will do the same for pitchers after.

```python
# build model
hitters_model = LogisticRegression(penalty = 'l1', random_state = rs)

X_train_PCA = PCA_hit.fit_transform(X_train)
X_test_PCA = PCA_hit.transform(X_test)
hitters_model.fit(X_train_PCA, y_train)

# cross-validation
scores = cross_val_score(estimator = hitters_model, X = X_train_PCA, y = y_train, cv = 10)
print('CV accuracy: %s' % scores)
print('CV accuracy range: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# output dataframe
output = pd.DataFrame(hitters_model.predict(X_test_PCA))
output.index = y_test.index
hitters_pred = pd.concat([y_test, output], axis = 1)
hitters_pred.rename({0: 'predicted'}, axis = 1, inplace = True)
hitters_pred.T
```

Overall, our hitting model does pretty well with predicting. There is evidence of a small bit of potential overfitting. Let's take the same reasoning and apply it quickly to the pitching dataset.

```python
# generally, we do pretty well with the prediction
hitters_model.score(X_test_PCA, y_test)
```

## Final predictive step: future Hall of Famers
Predicting the future: who will be inducted.

```python
# hitting predictions
active_hitters = players_active.merge(hitters, left_index = True, right_index = True)
hitters_adj = hitters.merge(players[['active', 'seasons_played']], left_index = True, right_index = True)
eligible_novoted = hitters_adj[(hitters_adj.active == 0) & (hitters_adj.seasons_played >= 10) & (hitters_adj.inducted == 0) & (hitters_adj['VOTE%'] > 0)].copy()
eligible_novoted.drop(labels = ['active'], axis = 1, inplace = True)

# add players for re-votes 
active_hitters = pd.concat([active_hitters, eligible_novoted], axis = 0)

# filter out players who have not yet hit 10 seasons
active_hitters = active_hitters[active_hitters.seasons_played >= 10]
print('Total number of active/eligible players with 10+ seasons: {}'.format(len(active_hitters)))

# drop unused columns
active_hitters.drop(labels = ['VOTE%', 'inducted', 'seasons_played', 'name', \
                              'CYA', '3000K', '300W', '300SV', 'so', 'g_idp',\
                              'cs', 'SAC'], axis = 1, inplace = True)

active_hitters.fillna(value = 0, inplace = True)
X_AH = active_hitters.copy()

# standard scaling
X_AH[hitting_scalable] = hit_scaler.transform(X_AH[hitting_scalable])
X_AH = PCA_hit.transform(X_AH)

# prediction
active_hitters['prediction'] = hitters_model.predict(X_AH)

# joining to dataframe of names
prediction_hit = players.merge(active_hitters[['prediction']], how = 'right', left_index = True, right_index = True)
prediction_hit.sort_values(by = 'prediction', inplace = True, ascending = False)

# full list of predictions
prediction_hit[prediction_hit.prediction == 1]
```
