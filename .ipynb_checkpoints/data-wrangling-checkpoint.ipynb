# MLB Hall of Fame

Usng Lahman's Baseball Database to predict which players will be inducted into the MLB Hall of Fame. This is for data wrangling; afterwards, we will run an analysis to predict future HOF'ers (grading against the 2015-2018 classes) and run some visualizations as well.

#### Dataset sources:   
Sean Lahman, Lahman's Baseball Database: [MLB_Kaggle](https://www.kaggle.com/seanlahman/the-history-of-baseball/data)

We limit ourselves to the following .csv files in this analysis:
- `all_star.csv`: All-Star appearances
- `batting.csv` and `pitching.csv`: Batting data and pitching data, used to evaluate player value for HOF candidacy
- `player_award_vote.csv`: MVP, Cy Young, and other votes
- `hall_of_fame.csv`: Response variable, induction into HOF
- `player.csv`: Dataset of player information to match up names, etc.

If possible, it would be interesting to incorporate the following data files in these ways:
- `postseason.csv`: World Series and postseason appearances per player
- `fielding.csv`, `fielding_outfield.csv`, `fielding_postseason.csv`, `batting_postseason.csv`, `pitching_postseason.csv`: Fielding and postseason data, similar to batting
- `manager.csv`: Managerial statistics to evaluate these members for Hall of Fame status
- Counting number of times leading the league in a stat

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data cleaning
Removing unnecessary rows/columns, cleaning up size of datasets. For the sake of dataset size, we remove the datasets that likely provide less relevance but have significantly higher impacts on the chance of a player being inducted into the Hall of Fame.

#### All-Star appearances
We don't care about the specific performance, per se, but rather people look at how many All-Star games a player was selected for.

```python
# counts of all-star performances
all_star = pd.read_csv('data/all_star.csv')
all_star = all_star.groupby('player_id')['year'].count().reset_index()
all_star.rename({'year': 'All-Star'}, axis = 1, inplace = True)
all_star.sort_values(by = 'All-Star', ascending = False).head()
```

#### Batting statistics
We want to look at batting statistics for players, but want to only look at those for which the number of at-bats is reasonable (otherwise, summary statistics makes no sense). Moreover, we only select players that played for more than 10 seasons, since that is a HOF eligibility requirement.

```python
# importing data
batting = pd.read_csv('data/batting.csv')

# counting number of seasons
total_seasons = batting.groupby('player_id')['year'].count().reset_index()
# year_check = total_seasons[total_seasons['year'] >= 10]
batting = batting.merge(total_seasons, on = 'player_id')

# save copy of dataframe
player_batting_full = batting.copy()

# drop extraneous rows and rename
player_batting_full.drop(labels = ['year_y'], axis = 1, inplace = True)
player_batting_full.rename({'year_x': 'year'}, axis = 1, inplace = True)
batting.drop(labels = ['year_x', 'year_y'], axis = 1, inplace = True)
batting.head()
```

```python
# aggregating batting statistics by player
batting = batting.groupby('player_id').sum().reset_index()
batting.drop(labels = ['stint'], axis = 1, inplace = True)
batting.head()
```

```python
# doing similarly for postseason batting
postseason_batting = pd.read_csv('data/batting_postseason.csv')
postseason_batting = postseason_batting.groupby('player_id').sum().reset_index()
postseason_batting.drop(labels = ['year'], axis = 1, inplace = True)
postseason_batting.head()
```

For players that had almost no postseason at-bats/games or appearances at the plate, we can assume they were either pitchers, or rarely qualified for the postseason. As such, it makes no sense to include them in the dataset. We can examine this by plotting frequencies of at-bats for each and distributions.

```python
plt.plot(batting['ab'])
plt.title('At-Bats in Regular Season')
plt.show()

plt.plot(postseason_batting['ab'])
plt.title('At-Bats in Postseason')
plt.show()
```

Oddly enough, we don't actually see much here. Let's look at the distributions instead, removing all of our zeroes first.

```python
# removing zeros
batting = batting[batting['ab'] != 0]
postseason_batting = postseason_batting[postseason_batting['ab'] != 0]

pd.DataFrame({'batting': batting['ab'], 'postseason': postseason_batting['ab']}).describe()
```

50% of the players in the dataset went to bat less than what is considered by MLB standards (3.1 PA/game) in the postseason for five full games (a standard series). So, let's remove them from consideration.

```python
# removing 50% of the dataset
postseason_ab_half = postseason_batting['ab'].quantile(0.5)
postseason_batting = postseason_batting[postseason_batting['ab'] >= postseason_ab_half]
```

Now let's aggregate some hitting statistics. Let's include IBBs/HBPs into the walks, and agggregate sac flies too.   

We'll capitalize all of our created features for ease of understanding/readability.

```python
# walks = walks + ibb + hbp
batting['BB'] = batting['bb'] + batting['ibb'] + batting['hbp']
batting['SAC'] = batting['sf'] + batting['sh']
batting.drop(labels = ['bb', 'ibb', 'hbp', 'sh', 'sf'], axis = 1, inplace = True)
```

```python
# average, total bases, XBH, OBP, slugging
batting['TB'] = batting['double'] + batting['triple']*2 + batting['hr']*3 + batting['h']
batting['XBH'] = batting['double'] + batting['triple'] + batting['hr']
batting['AVG'] = batting['h']/batting['ab']
batting['OBP'] = round((batting['h'] + batting['BB'])/(batting['ab'] + batting['BB'] + batting['SAC']), 3)
batting['SLG'] = round(batting['TB']/batting['ab'], 3)
batting['OPS'] = batting['OBP'] + batting['SLG']

batting['AVG'], batting['OBP'], batting['SLG'], batting['OPS'] = \
    round(batting['AVG'], 3), round(batting['OBP'], 3), round(batting['SLG'], 3), round(batting['OPS'], 3)

batting.head()
```

#### Fielding statistics

Removing players with less than 10 seasons of games worth played

```python
fielding = pd.read_csv('data/fielding.csv')
fielding = fielding.merge(total_seasons, on = 'player_id')
fielding.drop(labels = ['year_x', 'year_y'], axis = 1, inplace = True)

fielding = fielding.groupby('player_id').sum().reset_index()
fielding.drop(labels = ['stint'], axis = 1, inplace = True)
fielding.head()
```

Let's aggregate some statistics for fielding:

```python
# fielding percentage
fielding['FPCT%'] = (fielding['po'] + fielding['a'])/(fielding['po'] + fielding['a'] + fielding['e'])

# range factor
fielding['RF'] = 9*(fielding['po'] + fielding['a'])/(fielding['inn_outs'])

# steal percentage
fielding['CS%'] = fielding['cs']/(fielding['sb'] + fielding['cs'])
fielding.drop(labels = ['g', 'gs', 'pb', 'wp', 'sb', 'cs'], axis = 1, inplace = True)
```

```python
fielding.head()
```

#### Pitching data
Pitchers over 10 seasons

```python
pitching = pd.read_csv('data/pitching.csv')
pitching = pitching.merge(total_seasons, on = 'player_id')
pitching.drop(labels = ['year_x', 'year_y', 'gs', 'gf'], axis = 1, inplace = True)
pitching.head()
```

```python
# aggregating pitching statistics by player
pitching = pitching.groupby('player_id').sum().reset_index()
pitching.drop(labels = ['stint'], axis = 1, inplace = True)
pitching.head()
```

There are 24 columns here, so let's delete some of the ones that may not make sense, or combine them. For example: HBP, BB can go together, while sacrifices, runs allowed, GIDPs, Balks, probably aren't so relevant.

```python
pitching['BB'] = pitching['bb'] + pitching['ibb'] + pitching['hbp']
pitching.drop(labels = ['g_idp', 'r', 'bk', 'wp', 'ibb', 'bb', 'hbp', 'baopp', 'era'], axis = 1, inplace = True)
pitching.head()
```

Now let's calculate aggregate/summary statistics again:

```python
pitching['ERA'] = round(27*pitching['er']/pitching['ipouts'], 2)
pitching['WHIP'] = round(3*(pitching['h'] + pitching['BB'])/pitching['ipouts'], 3)
pitching['BAA'] = round(pitching['h']/(pitching['bfp'] - pitching['BB'] - pitching['sh'] - pitching['sf']), 3)
pitching['K/9'] = round(27*pitching['so']/pitching['ipouts'], 2)
pitching['BB/9'] = round(27*pitching['BB']/pitching['ipouts'], 2)
pitching['K/BB'] = round(pitching['so']/pitching['BB'], 2)
pitching.drop(labels = ['bfp', 'sh', 'sf'], axis = 1, inplace = True)
pitching.head()
```

## Creating the datasets for modeling

#### Hall of Fame
Creating both a Hall of Fame table and a table for players who are not yet eligible (playing w/in last 5 years). Only looking at HOF Players.

```python
# importing and cleaning dataset
hof = pd.read_csv('data/hall_of_fame.csv')
hof = hof[(hof['category'] == 'Player') & (hof['votedby'] == 'BBWAA')]
hof['VOTE%'] = hof['votes']/hof['ballots'] # percetange of required votes needed (cutoff = 75%)
hof['%REQ'] = hof['votes']/hof['needed'] # percentage relative to how many needed (>1 = HOF)
hof.drop(labels = ['votedby', 'needed', 'votes', 'ballots', 'needed_note', 'category'], axis = 1, inplace = True)
hof['inducted'] = hof['inducted'].apply(lambda x: 1 if x == 'Y' else 0) # numerical boolean
```

```python
# vote% and %req depend on the number needed and number of voters, so let's see if this is actually duplicatory
print(hof[(hof['VOTE%'] >= .75) ^ (hof['%REQ'] >= 1)].head())

# seeing none, we can remove the latter, and focus just on vote percentage
hof.drop(labels = ['%REQ'], axis = 1, inplace = True)
```

```python
# table of ineligible players - batting within 5 years
ineligible = player_batting_full[player_batting_full['year'] >= 2010].drop_duplicates(subset = 'player_id')
ineligible = ineligible[['player_id']]
ineligible['active'] = 1
```

#### Player data
Player names

```python
players = pd.read_csv('data/player.csv')
players = players[['player_id', 'name_first', 'name_last']]
players['name'] = players['name_first'] + " " + players['name_last']
players.drop(labels = ['name_first', 'name_last'], axis = 1, inplace = True)
players.head()
```

Number of triple-crowns (exceedingly rare)

```python
# player award dataset
player_award = pd.read_csv('data/player_award.csv')

# identifying triple crowns
pitching_tc = player_award[(player_award.award_id == 'Pitching Triple Crown')]
batting_tc = player_award[(player_award.award_id == 'Triple Crown')]
pitching_tc = pitching_tc.groupby('player_id')['award_id'].count().reset_index()
batting_tc = batting_tc.groupby('player_id')['award_id'].count().reset_index()
tc = pd.concat([batting_tc, pitching_tc])
tc.rename({'award_id': 'TripleCrown'}, axis = 1, inplace = True)
```

```python
# joining to players
master_df = players.merge(tc, how = 'left', on = 'player_id')
master_df = master_df.merge(all_star, how = 'left', on = 'player_id')
master_df.head()
```

MVPs, Cy Youngs - we don't care about rookie of the year.

```python
# MVP/Cy Young voting
award_voting = pd.read_csv('data/player_award_vote.csv')
award_voting = award_voting[award_voting.award_id != 'Rookie of the Year']
cya = award_voting[award_voting.award_id == 'Cy Young']
mvp = award_voting[award_voting.award_id == 'MVP']
```

```python
# determining winners and matching back to find who got the award
mvp_winner = mvp.groupby(['year', 'league_id'])['points_won'].max().reset_index()
mvp_awards = mvp.merge(mvp_winner, on = ['year', 'league_id', 'points_won'])
mvp_counts = mvp_awards.groupby('player_id')['year'].count().reset_index()
mvp_counts.rename({'year': 'MVP'}, axis = 1, inplace = True)

cya_winner = cya.groupby(['year', 'league_id'])['points_won'].max().reset_index()
cya_awards = cya.merge(cya_winner, on = ['year', 'league_id', 'points_won'])
cya_counts = cya_awards.groupby('player_id')['year'].count().reset_index()
cya_counts.rename({'year': 'CYA'}, axis = 1, inplace = True)
```

```python
# filling in table of master_df
master_df = master_df.merge(mvp_counts, how = 'left', on = 'player_id').merge(cya_counts, how = 'left', on = 'player_id')
master_df.fillna(value = 0, axis = 1, inplace = True)
master_df.head()
```

Add in all-time leaders, 3000 hit club, 500 HR club, 1500 RBI club, 3000 K club, 300 W club, 300 SV club, etc.

```python
# specific filters in cumulative batting
convert_bool = lambda x: 1 if x else 0

hit_3000 = pd.concat([batting.player_id, (batting.h >= 3000).apply(convert_bool)], axis = 1)
hr_500 = pd.concat([batting.player_id, (batting.hr >= 500).apply(convert_bool)], axis = 1)
rbi_1500 = pd.concat([batting.player_id, (batting.rbi >= 1500).apply(convert_bool)], axis = 1)

k_3000 = pd.concat([pitching.player_id, (pitching.so >= 3000).apply(convert_bool)], axis = 1)
w_300 = pd.concat([pitching.player_id, (pitching.w >= 300).apply(convert_bool)], axis = 1)
sv_300 = pd.concat([pitching.player_id, (pitching.sv >= 300).apply(convert_bool)], axis = 1)
```

```python
# adding to dataframe
for i in [hit_3000, hr_500, rbi_1500, k_3000, w_300, sv_300]:
    master_df = master_df.merge(i, how = 'left', on = 'player_id')
master_df.rename({'h': '3000H', 'hr': '500HR', 'rbi': '1500RBI', 'so': '3000K', 'w': '300W', 'sv': '300SV'}, axis = 1, inplace = True)
```

```python
# filling in other values
master_df.fillna(value = 0, axis = 1, inplace = True)
master_df.head()
```

```python
# determine if players are in the hall of fame yet
master_df = master_df.merge(hof, how = 'left', on = 'player_id')
master_df = master_df.merge(ineligible, how = 'left', on = 'player_id')
```

```python
# add number of seasons played
master_df = master_df.merge(total_seasons, how = 'left', on = 'player_id')
master_df.rename({'year': 'seasons_played'}, axis = 1, inplace = True)
```

```python
# clean master_df of year_id since the variable does not matter to us
master_df.drop('yearid', axis = 1, inplace = True)
master_df.fillna(value = 0, axis = 1, inplace = True)
```

## Output clean data
We have batters, pitchers, and the master_df with elite honors + hall of fame + eligibility status.

```python
# rearrange columns
cols = list(master_df)
# new_cols = [cols[0], cols[1], cols[-2], cols[-1]] + cols[2:-4] + [cols[-3], cols[-4]]
new_cols = ['player_id','name','active','seasons_played','All-Star','TripleCrown','MVP','CYA','3000H','500HR','1500RBI','3000K','300W','300SV','VOTE%','inducted']
master_df = master_df[new_cols]
```
