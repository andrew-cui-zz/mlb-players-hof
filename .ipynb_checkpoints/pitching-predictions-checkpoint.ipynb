# MLB Hall of Fame

Usng Lahman's Baseball Database to predict which players will be inducted into the MLB Hall of Fame. Predictive analytics of hall of fame members.

#### Dataset sources:   
Sean Lahman, Lahman's Baseball Database: [MLB_Kaggle](https://www.kaggle.com/seanlahman/the-history-of-baseball/data)

We limit ourselves to the following .csv files in this analysis:
- `all_star.csv`: All-Star appearances
- `batting.csv` and `pitching.csv`: Batting data and pitching data, used to evaluate player value for HOF candidacy
- `player_award_vote.csv`: MVP, Cy Young, and other votes
- `hall_of_fame.csv`: Response variable, induction into HOF
- `player.csv`: Dataset of player information to match up names, etc.

```python
# base python packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# sklearn pre-processing
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

# sklearn model selection
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.pipeline import Pipeline

# sklearn models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression

# block warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# import datasets
pitching_hof = pd.read_csv('model_data/pitching_hof.csv')
players_active = pd.read_csv('model_data/players_active.csv')
hof_eligible = pd.read_csv('model_data/hof_eligible.csv')
hof_induction = pd.read_csv('model_data/hof_induction.csv')
pitchers = pd.read_csv('model_data/pitchers.csv')
players = pd.read_csv('model_data/players_df.csv')

pitching_hof.set_index('player_id', inplace = True)
players_active.set_index('player_id', inplace = True)
hof_eligible.set_index('player_id', inplace = True)
hof_induction.set_index('player_id', inplace = True)
pitchers.set_index('player_id', inplace = True)
players.set_index('player_id', inplace = True)

pitching_hof.head()
```

## Scaling and initial tests of data - pitching
We use many of the similar approaches to scaling the data from the hitting model. It would make sense for the two voting processes to be fundamentally similar, even if different factors are considered.

```python
rs = 19
```

```python
# preparing dataset
X = pitching_hof.copy()
X['All-Star'] = X['All-Star']/X['seasons_played']
X.drop(labels = ['seasons_played', 'inducted', '3000H', '500HR', '1500RBI', 'hr', 'er', 'ipouts', 'g', 'BB', 'h'], axis = 1, inplace = True)
y = pitching_hof[['inducted']].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = rs)
```

```python
# standard scaling, neglecting our cols that do not need scaling due to their binary or percentage nature
pitching_cols = list(X_train)
pitching_scalable = ['All-Star'] + pitching_cols[7:]

# partially standard scaling
def standardscaling(X_train, X_test, scalable):
    scaler = StandardScaler()
    X_train[scalable] = scaler.fit_transform(X_train[scalable])
    X_test[scalable] = scaler.transform(X_test[scalable])
    return X_train, X_test, scaler

X_train, X_test, pitch_scaler = standardscaling(X_train, X_test, pitching_scalable)
X_train.head()
```

```python
# plot learning curve
pipe_lr = Pipeline([('clf', LogisticRegression())])
train_sizes, train_scores, test_scores = \
                learning_curve(estimator = pipe_lr,
                        X = X_train, y = y_train,
                        train_sizes = np.linspace(0.2, 1.0, 8), 
                        cv = 10)

# learning curve ranges
def learningcurve_ranges(train_scores, test_scores):
    train_mean = np.mean(train_scores, axis = 1)
    train_std = np.std(train_scores, axis = 1)
    test_mean = np.mean(test_scores, axis = 1)
    test_std = np.std(test_scores, axis = 1)
    return train_mean, train_std, test_mean, test_std

# plotting learning curve
def plot_learningcurve(train_sizes, train_scores, test_scores, model_name):
    train_mean, train_std, test_mean, test_std = learningcurve_ranges(train_scores, test_scores)
    plt.plot(train_sizes, train_mean, 'b', marker='o')
    plt.plot(train_sizes, test_mean, 'r', marker='o')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color = 'red', alpha = 0.1)
    
    plt.title('Learning curve, {}'.format(model_name))
    plt.legend(('training', 'test'), fancybox = True)
    plt.show()

plot_learningcurve(train_sizes, train_scores, test_scores, 'LogisticRegression')
```

Here, overfitting seems to be less of a problem than in the previous case, but we'll still look at PCA and find around the 98% of explained variance we did last time for parsimony.

## PCA
Dimensionality reduction of the data

```python
# let's determine optimal number of components
pca_model = PCA(n_components = X_train.shape[1]) # all components
X_PCA = pca_model.fit_transform(X_train)

variance = pca_model.explained_variance_ratio_ #variance ratios                                        

# cumulative sum of variance, for the first [n] features
cv = np.cumsum(np.round(pca_model.explained_variance_ratio_, decimals = 3)*100)
```

```python
# plot cv vs features
plt.plot(cv, c = 'black')
plt.xlabel('number of features')
plt.ylabel('cumulative sum variance')
plt.xticks(np.arange(0, len(pitching_cols), 1))
plt.title('cv of pca = {}'.format(len(pitching_cols)))
plt.axhline(y = 99, color = 'purple', alpha = .75)
plt.axhline(y = 98, color = 'green', alpha = .75)
plt.axhline(y = 95, color = 'orange', alpha = .75)
plt.axhline(y = 90, color = 'yellow', alpha = .75)
ax = plt.axes()        
ax.xaxis.grid(True, alpha = 0.5) 
plt.show()
```

Our PCA cutoffs are similar, withh `n_components = 6` giving us 95% variance explained.

```python
# PCA model
PCA_pitchers = PCA(n_components = 6)
X_train_PCA = PCA_pitchers.fit_transform(X_train)
X_test_PCA = PCA_pitchers.transform(X_test)

# rerunning logistic regression 
model = LogisticRegression(random_state = rs)
model.fit(X_train_PCA, y_train)
print('Model score: {}'.format(model.score(X_test_PCA, y_test)))

# cross-validation
scores = cross_val_score(estimator = model, X = X_train_PCA, y = y_train, cv = 10)
print('CV accuracy: %s' % scores)
print('CV accuracy range: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))

# plot learning curve
pipe_lr = Pipeline([('clf', model)])
train_sizes, train_scores, test_scores = \
                learning_curve(estimator = pipe_lr,
                        X = X_train_PCA, y = y_train,
                        train_sizes = np.linspace(0.2, 1.0, 8), 
                        cv = 10)
plot_learningcurve(train_sizes, train_scores, test_scores, 'LogisticRegression')
```

This actually minimizes overfitting even better than the hitting dataset, although that's not surprising given that pitchers often have less metrics to be evaluated on that can differ so much (ex: both power and finesse pitchers will have lower ERAs and higher win counts, if they make it to the hall of fame).

## Modeling and final predictions - pitching
We will run some of the models again, but here will limit us just to the best algorithms from the hitting perspective - ideally, we want a model that is generalizable.

Schema for `classifier_results` to compare different predictive algorithms copied from HW5 in [CIS 545: Big Data Analytics](https://www.cis.upenn.edu/~cis545/)

```python
# preparing dataset
X = pitching_hof.copy()
X['All-Star'] = X['All-Star']/X['seasons_played']
X.drop(labels = ['seasons_played', 'inducted', '3000H', '500HR', '1500RBI', 'hr', 'er', 'ipouts', 'g', 'BB', 'h'], axis = 1, inplace = True)
y = pitching_hof[['inducted']].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = rs)

# classifier results dictionary
classifier_results = []
```

```python
# baseline: 96.57%
print(y.groupby('inducted')['inducted'].count())
classifier_results.append({'Classifier': 'Everything = 0', 'Score': (1166 - 40)/1166})
(1166 - 40)/1166
```

```python
# standard scaling and PCA
X_train, X_test, pitch_scaler = standardscaling(X_train, X_test, pitching_scalable)
X_train_PCA = PCA_pitchers.fit_transform(X_train)
X_test_PCA = PCA_pitchers.transform(X_test)
```

```python
# helper function for classification
def model_score(model):
    model.fit(X_train_PCA, y_train)
    test_score = model.score(X_test_PCA, y_test)
    return test_score
```

```python
# decision trees for various heights
for i in range(4, 8, 2):
    dt_model = DecisionTreeClassifier(max_depth = i, random_state = rs)
    classifier_results.append({'Classifier': 'DecTree', 'Depth': i, 'Score': model_score(dt_model)})

for i in [10, 15, 20]:
    rf_model = RandomForestClassifier(n_estimators = i, random_state = rs)
    classifier_results.append({'Classifier': 'RandomForest', 'Score': model_score(rf_model), 'Count': len(rf_model.estimators_)})
```

Boost and Bagging classifiers.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier

## Bagging classifier Decision Tree, depths varied
for i in range(4, 8, 2):
    dt_model = DecisionTreeClassifier(max_depth = i, random_state = rs) # 2 depth
    dt_bag = BaggingClassifier(base_estimator = dt_model, n_estimators = 5, random_state = rs)
    classifier_results.append({'Classifier': 'Bag-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

    dt_bag = BaggingClassifier(base_estimator = dt_model, n_estimators = 10, random_state = rs)
    classifier_results.append({'Classifier': 'Bag-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

    dt_bag = AdaBoostClassifier(base_estimator = dt_model, n_estimators = 5, random_state = rs)
    classifier_results.append({'Classifier': 'Boost-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

    dt_bag = AdaBoostClassifier(base_estimator = dt_model, n_estimators = 10, random_state = rs)
    classifier_results.append({'Classifier': 'Boost-DecTree', 'Depth': i, 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

## Boost classifier Decision Tree
for i in [5, 10, 15, 20]:
    dt_boost = AdaBoostClassifier(base_estimator = dt_model, n_estimators = i, random_state = rs)
    classifier_results.append({'Classifier': 'Boost-DecTree', 'Score': model_score(dt_boost), 'Count': len(dt_boost.estimators_)})
```

Let's view our final classifiers and select a model.

```python
# output classifiers as a dataframe
pitching_classifiers = pd.DataFrame(classifier_results)

# baseline score
baseline = pitching_classifiers.iloc[0, 3]
pitching_improved = pitching_classifiers[pitching_classifiers['Score'] > baseline]

# models
pitching_classifiers.sort_values(by = 'Score', inplace = True, ascending = False)
pitching_classifiers
```

Similarly, let's use a Decision Tree or Random Forest and compare the different options. An SVM loses evidently here.

```python
tree_select = []
dt_model = DecisionTreeClassifier(max_depth = 4, random_state = rs)

for i in range(1, 5): 
    dt_bag = BaggingClassifier(base_estimator = dt_model, n_estimators = i, random_state = rs)
    tree_select.append({'Classifier': 'Bag-DecTree', 'Score': model_score(dt_bag), 'Count': len(dt_bag.estimators_)})

    dt_boost = AdaBoostClassifier(base_estimator = dt_model, n_estimators = i, random_state = rs)
    tree_select.append({'Classifier': 'Boost-DecTree', 'Score': model_score(dt_boost), 'Count': len(dt_boost.estimators_)})

    
for i in range(1, 10): 
    rf_model = RandomForestClassifier(n_estimators = i, random_state = rs)
    tree_select.append({'Classifier': 'RandomForest', 'Score': model_score(rf_model), 'Count': len(rf_model.estimators_)})

tree_df = pd.DataFrame(tree_select)
tree_df.drop_duplicates(subset = ['Score', 'Count'], inplace = True)
print(tree_df.groupby(['Classifier'])['Score'].max())

# maximum is DecisionTree Bag
max_hitting = tree_df.groupby(['Classifier'])['Score'].max()[0]
tree_df[tree_df.Score == max_hitting]
```

## Final model: pitchers
For the sake of simplicity here, we select the single-bagged Decision Tree level-4. This gives us a consistent story with the modeling from the previous section.         

The intuitive story based on this is that there may be several "planes" of separation between the great and regular players, and a mix of performance on those aspects can be featured into the actual hall of fame - that's pretty reasonable, so we'll stick with it. One interesting thing to note is how we only needed 6 PCA features this time.

```python
# build model
base_tree = DecisionTreeClassifier(max_depth = 4, random_state = rs)
pitchers_model = BaggingClassifier(base_estimator = base_tree, n_estimators = 1, random_state = rs)

X_train_PCA = PCA_pitchers.fit_transform(X_train)
X_test_PCA = PCA_pitchers.transform(X_test)
pitchers_model.fit(X_train_PCA, y_train)

# cross-validation
scores = cross_val_score(estimator = pitchers_model, X = X_train_PCA, y = y_train, cv = 10)
print('CV accuracy: %s' % scores)
print('CV accuracy range: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# output dataframe
output = pd.DataFrame(pitchers_model.predict(X_test_PCA))
output.index = y_test.index
pitchers_pred = pd.concat([y_test, output], axis = 1)
pitchers_pred.rename({0: 'predicted'}, axis = 1, inplace = True)
pitchers_pred.T 
```

Overall model score is pretty good.

```python
# generally, we do pretty well with the prediction
pitchers_model.score(X_test_PCA, y_test)
```

## Final predictive step: future Hall of Famers
Predicting the future: who will be inducted.

```python
# pitching predictions
active_pitchers = players_active.merge(pitchers, left_index = True, right_index = True)
pitchers_adj = pitchers.merge(players[['active', 'seasons_played']], left_index = True, right_index = True)
eligible_novoted = pitchers_adj[(pitchers_adj.active == 0) & (pitchers_adj.seasons_played >= 10) & (pitchers_adj.inducted == 0) & (pitchers_adj['VOTE%'] > 0)].copy()
eligible_novoted.drop(labels = ['active'], axis = 1, inplace = True)

# add players for re-votes 
active_pitchers = pd.concat([active_pitchers, eligible_novoted], axis = 0)

# filter out players who have not yet hit 10 seasons
active_pitchers = active_pitchers[active_pitchers.seasons_played >= 10]
print('Total number of active/eligible players with 10+ seasons: {}'.format(len(active_pitchers)))

# drop unused columns
active_pitchers['All-Star'] = active_pitchers['All-Star']/active_pitchers['seasons_played']
active_pitchers = active_pitchers[pitching_cols]

active_pitchers.fillna(value = 0, inplace = True)
X_AP = active_pitchers.copy()

# standard scaling
X_AP[pitching_scalable] = pitch_scaler.transform(X_AP[pitching_scalable]) ####
X_AP = PCA_pitchers.transform(X_AP)

# prediction
active_pitchers['prediction'] = pitchers_model.predict(X_AP)

# joining to dataframe of names
prediction_pitch = players.merge(active_pitchers[['prediction']], how = 'right', left_index = True, right_index = True)
prediction_pitch.sort_values(by = 'prediction', inplace = True, ascending = False)

# predictions
prediction_pitch[prediction_pitch.prediction == 1]
```
